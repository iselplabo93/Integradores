```python
from pylab import*
from scipy.integrate import*
%matplotlib inline
```

```python
#oscilador armonico y''+w_0^2y=0, con y'=v, v'=-w_0^2, y_v[0]=y,y_v[1]=v
def osc(y_v,t,omega2=1):
    dy=y_v[1]
    dv=-omega2*y_v[0]
    return array([dy,dv])
```

```python
t=linspace(0,10)
inicio=[0,1]
sol=odeint(osc,inicio,t)
plot(sol[:,0],sol[:,1])
```

```python
#con friccion
#oscilador armonico y''+w_0^2y=0, con y'=v, v'=-w_0^2, y_v[0]=y,y_v[1]=v
def osc2(y_v,t,omega2=1,mu=2):
    dy=y_v[1]
    dv=-omega2*y_v[0]-mu*y_v[1]
    return array([dy,dv])
```

```python
t=linspace(0,100,1000)
inicio=[0,1]
sol=odeint(osc2,inicio,t,args=(1,0.1))
xlim([-1,1]),ylim([-1,1]) #funciona como funcion, argumentos de una funcion en solo paquete
plot(sol[:,0],sol[:,1])

```

```python
fig=figure() #crea una figura, y plot genera una figura donde es una variable, al ejecutar muestra la memoria donde lo guardo
ejes=axes(xlim=(-1,1),ylim=(-1,1)) #limites asociados a axes, argumentos de funciones
trayectoria,=ejes.plot([],[]) #la coma es para que se comporte como apuntador, a ejes pegarle una grafica
punto,=ejes.plot([],[],'o')
tache,=ejes.plot([0],[0],'x')
def init():
    trayectoria.set_data([],[]) #caen datos vacios, funcion de inicio-datos nulos
    punto.set_data([],[])
    return trayectoria,punto,
def el_movimiento(i):#hacer dibujos
    y,v=sol[:,0],sol[:,1]
    trayectoria.set_data(y[:i+1],v[:i+1]) #pasa datos a traves de la trayectoria y los manda al plot
    punto.set_data(y[i],v[i])
    return trayectoria,punto,
animation.FuncAnimation(fig,el_movimiento,init_func=init,frames=100) # entorno, con que hacer dibujos, de donde partir "borron y cuenta nueva"


```
