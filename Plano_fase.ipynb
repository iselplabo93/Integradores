```python
from pylab import *
%run 'Integradores.ipynb'
%matplotlib inline
```

```python
def osc(y,t,w02=1): #oscilador armonico-plano fase y'+w0^2sin(y)=g(t), y'=u, u'=-w0^2*sin(y)+g(t)
    return array([y[1],-w02*sin(y[0])]+0.1*sin(t)) # homogenea array([y[1],-w02*sin(y[0])]), inhomogenea array([y[1],-w02*sin(y[0])]+0.1*sin(t))
```

```python
for i in range(40): # range indica el numero de trayectorias
    y0,u0= pi*(2*rand()-1),0   #condiciones iniciales aleatorias, con u0=0
    t,sol=rk4(osc,[y0,u0],[0.,10.])# [10 indica tiempo], [1.,0.] indica condicion inicial
    plot(sol[:,0],sol[:,1])
zoom=pi
xlim([-zoom,zoom]),ylim([-zoom,zoom])
axes().set_aspect('equal'),grid(True)
show()

```

```python
y0,u0= 1.2,0   #condiciones iniciales
t,sol=rk4(osc,[y0,u0],[0.,1000.])# [10 indica tiempo], [1.,0.] indica condicion inicial
plot(sol[:,0],sol[:,1])
zoom=pi
xlim([-zoom,zoom]),ylim([-zoom,zoom])
axes().set_aspect('equal'),grid(True)
show()

```
