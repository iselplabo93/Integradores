```python
from pylab import *
%matplotlib inline
```

```python
def f(x,omega=1):
    return sin(omega*x)
def g(h_de_x,x,args=[]):
    if len(args)==0:
        resultado=exp(-x)*h_de_x(x)
    else :
        resultado=exp(-x)*h_de_x(x,args)
    return resultado
```

```python
x=linspace(0,2*pi,100)
for i in range(5):
    plot(x,g(f,x,[i]))
show()
```
